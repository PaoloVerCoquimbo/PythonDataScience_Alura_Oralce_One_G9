### Valor promedio por tipo de propiedad

```python
# Con groupby junta  los datos por tipo con valor solo se toma el promedio con el metodo mean de la columna Valor
# con esto se usa solo los datos numericos datos.groupby('Tipo').mean(numerionly)

datos.groupby('Tipo')['Valor'].mean()
```

```python
# se va organizar tipo DataFrame y ordenardo por valor desecendente
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
#
```

```python
df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
# kind barras orizontales  , figsize tamaño de la figura  , color
df_tipo_precio.plot(kind='barh',figsize=(20,9), color='purple')
```
